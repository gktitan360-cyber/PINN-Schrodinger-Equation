```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 1.0
N = 100
dx = L / (N + 1)

# Hamiltonian matrix
H = np.zeros((N, N))

for i in range(N):
    H[i, i] = 1 / dx**2

    if i > 0:
        H[i, i-1] = -1 / (2 * dx**2)

    if i < N-1:
        H[i, i+1] = -1 / (2 * dx**2)

# Solve eigenvalue problem
E, psi = np.linalg.eigh(H)

# Print ground state energy
print("Ground State Energy:", E[0])

# Plot ground state
x = np.linspace(0, L, N)

plt.figure(figsize=(6,4))
plt.plot(x, abs(psi[:,0]), linewidth=2)
plt.xlabel("x")
plt.ylabel(r"$\psi(x)$")
plt.title("Ground State of the Infinite Quantum Well")
plt.grid(True)
plt.show()

print("First Excited State Energy:", E[1])
plt.figure(figsize=(6,4))
plt.plot(x, psi[:,1], linewidth=2)
plt.xlabel("x")
plt.ylabel(r"$\psi(x)$")
plt.title("First Excited State of the Infinite Quantum Well")
plt.grid(True)
plt.show()

print("Second Excited State Energy:", E[2])
plt.figure(figsize=(6,4))
plt.plot(x, psi[:,2], linewidth=2)
plt.xlabel("x")
plt.ylabel(r"$\psi(x)$")
plt.title("Second Excited State of the Infinite Quantum Well")
plt.grid(True)
plt.show()
```

```python
#Analytical method for Fintie Quantum well
import numpy as np
import matplotlib.pyplot as plt

N = 500

x_min = -5
x_max = 5

x = np.linspace(x_min, x_max, N)

dx = x[1] - x[0]

V0 = 20.0
a = 1.0

V = np.where(np.abs(x) < a, 0.0, V0)

main_diag = 2*np.ones(N)/dx**2
off_diag = -1*np.ones(N-1)/dx**2
T = (np.diag(main_diag)+ np.diag(off_diag,1)+ np.diag(off_diag,-1))
H = T + np.diag(V)

eigenvalues, eigenvectors = np.linalg.eigh(H)
E0 = eigenvalues[0]
psi0 = eigenvectors[:,0]
norm = np.sqrt(np.sum(psi0**2)*dx)
psi0 = psi0/norm

print("Ground-state energy =", E0)

plt.plot(x, psi0, label="Ground state")
plt.plot(x,V/V0*np.max(np.abs(psi0)),'--',label="Potential")
plt.xlabel("x")
plt.ylabel(r"$\psi(x)$")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 200
xmin = -4
xmax = 4

x = np.linspace(xmin, xmax, N)
dx = x[1] - x[0]

H = np.zeros((N, N))

for i in range(N):

    # Kinetic energy
    H[i, i] = 1/dx**2

    if i > 0:
        H[i, i-1] = -1/(2*dx**2)

    if i < N-1:
        H[i, i+1] = -1/(2*dx**2)

    # Harmonic oscillator potential
    H[i, i] += 0.5 * x[i]**2

E, psi = np.linalg.eigh(H)

print(f"Ground State Energy = {E[0]:.6f}")

psi0 = psi[:, 0]
psi0 /= np.sqrt(np.sum(psi0**2) * dx)

plt.figure(figsize=(6,4))
plt.plot(x, psi0, linewidth=2)
plt.xlabel("Position, x")
plt.ylabel(r"$\psi_0(x)$")
plt.title("Ground State of the Quantum Harmonic Oscillator")
plt.grid(True)
plt.show()
```
